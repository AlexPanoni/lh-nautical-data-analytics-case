# Convertendo arquivos JSON para CSV

```python
# Convertendo o arquivo JSON "clientes_crm" para o arquivo csv "customers_raw"

with open(r'JSON\clientes_crm.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

dfcustomers = pd.DataFrame(data)

dfcustomers.to_csv('customers_raw.csv', index=False)
```

```python
# Convertendo o arquivo JSON "custos_importacao" para o arquivo csv "costs_raw"

rows = []

with open(r'JSON\custos_importacao.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

for product in data:
    for hist in product['historic_data']:
        rows.append({
            'product_id': product['product_id'],
            'product_name': product['product_name'],
            'category': product['category'],
            'start_date': hist['start_date'],
            'usd_price': hist['usd_price']
        })

dfcosts = pd.DataFrame(rows)

dfcosts.to_csv('costs_raw.csv', index=False)
```

# Checagem inicial de problemas

```python
# Checando o CSV "produtos_raw"

df_prod = pd.read_csv('produtos_raw.csv')

df_prod.head()
df_prod.info()
df_prod.isnull().sum()
df_prod.duplicated().sum()
df_prod['actual_category'].value_counts().head(20)
```

**Problema**: Na prática, apenas 3 categorias. Escritas de diversas maneiras, sem padronização.

```python
df_prod[df_prod.duplicated()]
```

```python
df_prod['name'].value_counts().head(10)
```

**Problema**: Nomes de produtos duplicados. Não aparecem como linha duplicada pois apresentam diferenciação na forma como a categoria está descrita na última coluna.

```python
df_sales = pd.read_csv('vendas_2023_2024.csv')

df_sales.head()
df_sales.info()
df_sales.isnull().sum()
df_sales.duplicated().sum()
```

**Problemas**:   
        1 - Apenas 9895 linhas não nulas, embora os IDs vão até 9999.   
        2 - Coluna de data em formato string.

```python
# Checando datas

df_sales['sale_date'].head(20)
```

```python
pd.to_datetime(df_sales['sale_date'], errors='coerce')
```

```python
df_sales[pd.to_datetime(df_sales['sale_date'], errors='coerce').isna()]
```

**Problema**: Apenas 4982 das 9894 (~50%) datas estão em um formato reconhecível padrão.

```python
# Checando datas com o formato invertido

pd.to_datetime(df_sales['sale_date'], errors='coerce', dayfirst=True)
```

```python
df_sales[pd.to_datetime(df_sales['sale_date'], errors='coerce', dayfirst=True).isna()]
```

=> Aumento de 4982 para 7945 datas válidas

```python
df_sales['sale_date'].sample(20)
```

```python
df_sales['sale_date'].str.len().value_counts()
```

=> Datas estão ou no formato YYYY-MM-DD ou DD-MM-YYYY
